```python
from scipy import sparse
 # creating 2d array with 1s in diagnoal and 0s in other using eye function
eye = np.eye(4)
print("numpy array:\n", format(eye))
```

```python
# Convert the NumPy array to a SciPy sparse matrix in CSR format
# Only the nonzero entries are stored
sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy sparse CSR matrix:\n{}".format(sparse_matrix))
```

```python
# Step 1: Define the data values for the non-zero elements
# We're using `np.ones(4)` to create an array of 4 ones → [1., 1., 1., 1.]
data = np.ones(4)

# Step 2: Define the row indices of the non-zero elements
# These are the row positions of the ones in the matrix → [0, 1, 2, 3]
row_indices = np.arange(4)

# Step 3: Define the column indices of the non-zero elements
# Same as row indices here → [0, 1, 2, 3]
col_indices = np.arange(4)

# Step 4: Create a sparse matrix in COOrdinate (COO) format
# This constructs a 4x4 identity matrix where only the diagonal elements are 1
# The syntax is: coo_matrix((data, (rows, columns)), shape=(n_rows, n_cols))
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))

# Step 5: Print the resulting sparse matrix in COO format
print("COO representation:\n{}".format(eye_coo))
```
